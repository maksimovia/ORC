```python
import CoolProp.CoolProp as CP
from openpyxl import load_workbook
import pandas as pd

class Turbine():
    def __init__(self,P_end,stream):
        self.__KPDoi = 0.9
        self.__P_end= P_end
        self.__stream = stream
        
    def calc_turbine(self):
        s_end = self.__stream['s']
        h_end_teor = CP.PropsSI('H','P',self.__P_end,'S',s_end,fluid)
        self._h_end = self.__stream['h'] - self.__KPDoi*(self.__stream['h']-h_end_teor)
        t_end = CP.PropsSI('T','P',self.__P_end,'H',self._h_end,fluid)
        s_end = CP.PropsSI('S','P',self.__P_end,'H',self._h_end,fluid)
            
        stream = {'P': self.__P_end,'T':t_end,'fluid':fluid,'h':self._h_end,'s':s_end, 'G': G1} #поток переопределяется в каждом классе (параметры на выходе из турбины)
        # P = self.__P_end
        return stream
    # print(stream)
    # print(P)
   
    
    def Turbine_calc_energy (self):
        A_turbine = self.__stream['G'] *(self.__stream['h'] - self._h_end )
        return A_turbine
    

class Condensator():
    def __init__(self,T_end,stream):
        self.__T_end= T_end
        self.__stream = stream
        self.__dp = 0.99 
    
    def Condensator_calc(self):
        self._h_end = CP.PropsSI('H','P',self.__stream['P']*self.__dp,'T',self.__T_end,fluid)
        P_end = self.__stream['P']*self.__dp
        s_end = CP.PropsSI('S','P',self.__stream['P']*self.__dp,'T',self.__T_end,fluid)
        
        stream = {'P': P_end,'T':self.__T_end,'fluid':fluid,'h':self._h_end,'s':s_end, 'G': G1}
        return stream
    
    def Condensator_calc_energy (self):
        A_condenser = self.__stream['G'] *(self.__stream['h'] - self._h_end)
        return A_condenser
    
class Nasos():
    def __init__(self,P_end,stream):
        self.__P_end= P_end
        self.__stream = stream
        self.__KPDoi = 0.85
    
    def Nasos_calc(self):
        h_end_teor = CP.PropsSI('H','P',self.__P_end,'S',self.__stream['s'],fluid)
        self._h_end = self.__stream['h'] - (self.__stream['h']-h_end_teor)*self.__KPDoi
        T_end = CP.PropsSI('T','P',self.__P_end,'H',self._h_end,fluid)
        s_end = CP.PropsSI('S','P',self.__P_end,'H',self._h_end,fluid)
        
        stream = {'P': self.__P_end,'T':T_end,'fluid':fluid,'h':self._h_end,'s':s_end, 'G': G1}
        return stream 
    
    def Nasos_calc_energy (self):
        A_nasos = self.__stream['G'] *(self._h_end - self.__stream['h'])
        return A_nasos
        

class TO():
    def __init__(self, stream1,stream2):
        self.__stream1 = stream1
        self.__stream2 = stream2
        self.__dp1 = 0.99
        self.__dp2 = 0.99
    
    def TO_calc(self,Nedogrev):
        
        T_end1 = self.__stream2['T']-Nedogrev #допущения: недогрев на горячем конце, противоток
        P_end1 = self.__dp1*self.__stream1['P']
        self._h_end1 = CP.PropsSI('H','P',P_end1,'T',T_end1,fluid)
        s_end1 = CP.PropsSI('S','P',P_end1,'T',T_end1,fluid)
        G1 = self.__stream1['G']
        
        P_end2 = self.__dp2*self.__stream2['P']
        
        self._h_end2 = self.__stream2['h']-(self.__stream1['G'] *(self._h_end1 - self.__stream1['h'])/self.__stream2['G'])
#         T_end2 = CP.PropsSI('T','P',P_end2,'H',self._h_end2,fluid)
        T_end2 = 80+273.15
        s_end2 = CP.PropsSI('S','P',P_end2,'H',self._h_end2,fluid)
        G2 = self.__stream2['G']
        
        stream1 = {'P': P_end1,'T':T_end1,'fluid':fluid,'h':self._h_end1,'s':s_end1, 'G': G1} # холодный поток
        stream2 = {'P': P_end2,'T':T_end2,'fluid':fluid,'h':self._h_end2,'s':s_end2, 'G': G2} # горячий поток
        
        return stream1, stream2
    
    def TO_calc_energy (self):
        A_TO1 = self.__stream1['G'] *(self._h_end1 - self.__stream1['h'])
        A_TO2 = self.__stream2['G'] *(self.__stream2['h']- self._h_end2)
        return A_TO1, A_TO2

class Regenerator():
    def __init__(self, stream1,stream2):
        self.__stream1 = stream1 #горячий поток
        self.__stream2 = stream2 #холодный поток
        self.__dp1 = 0.99
        self.__dp2 = 0.99
    
    def Reg_calc (self, Nedogrev):
        T_end1 = self.__stream2['T']+Nedogrev #допущения: недогрев на горячем конце, противоток
        P_end1 = self.__dp1*self.__stream1['P']
        self._h_end1 = CP.PropsSI('H','P',P_end1,'T',T_end1,fluid)
        s_end1 = CP.PropsSI('S','P',P_end1,'T',T_end1,fluid)
        G1 = self.__stream1['G']
        
        P_end2 = self.__dp2*self.__stream2['P']
        
        self._h_end2 = self.__stream2['h']+(self.__stream1['G'] *(self.__stream1['h'] - self._h_end1)/self.__stream2['G'])
        T_end2 = CP.PropsSI('T','P',P_end2,'H',self._h_end2,fluid)
        s_end2 = CP.PropsSI('S','P',P_end2,'H',self._h_end2,fluid)
        G2 = self.__stream2['G']
        
        stream1 = {'P': P_end1,'T':T_end1,'fluid':fluid,'h':self._h_end1,'s':s_end1, 'G': G1} #горячий поток
        stream2 = {'P': P_end2,'T':T_end2,'fluid':fluid,'h':self._h_end2,'s':s_end2, 'G': G1} #холодный поток 
        
        return stream1, stream2
    
    def Reg_calc_energy (self):
        A_Reg1 = self.__stream1['G'] *(self.__stream1['h'] - self._h_end1)
        A_Reg2 = self.__stream2['G'] *(self._h_end2 - self.__stream2['h'])
        return A_Reg1, A_Reg2




#исходные данные, стартуем от входа в утилизатор
G1=3 #холодный поток
G2=3 #горячий поток
P=24*1e6
T=302.46
P2=1e5
T2=500 
nedogrev=20
fluid='CO2'
fluid2 = 'Water'
stream_ytil_turb = {'P':P,'T':T2-nedogrev,'fluid':fluid,'h':CP.PropsSI('H','T',T2-nedogrev,'P',P,fluid),'s':CP.PropsSI('S','T',T2-nedogrev,'P',P,fluid), 'G': G1}

stream_vnesh = {'P':P2,'T':T2,'fluid':fluid,'h':CP.PropsSI('H','T',T2,'P',P2,fluid),'s':CP.PropsSI('S','T',T2,'P',P2,fluid), 'G': G2}
#--------------------------------------------------------------------------------
P_end = 7*1e6
T_cond=15+273.15

stream_cond_nasos = {'P':P_end,'T':T_cond,'fluid':fluid,'h':CP.PropsSI('H','T',T_cond,'P',P_end,fluid),'s':CP.PropsSI('S','T',T_cond,'P',P_end,fluid), 'G': G1}

#


# print(stream_reg_ytil, 'stream_reg_ytil')         
# Ytil=TO(stream_reg_ytil,stream_vnesh) #init
# stream_ytil_turb = Ytil.TO_calc(20)[0]  #calc 
     
turb = Turbine(P_end,stream_ytil_turb)    
stream_turb_reg = turb.calc_turbine()

Nasoss = Nasos(24*1e6, stream_cond_nasos) 
stream_nasos_reg = Nasoss.Nasos_calc()

Regen = Regenerator(stream_turb_reg, stream_nasos_reg) #init

stream_reg_cond = Regen.Reg_calc(5)[0]  #calc, горячий 
stream_reg_ytil= Regen.Reg_calc(5)[1]  #calc, холодный



Ytil=TO(stream_reg_ytil,stream_vnesh) #init
stream_ytil_turb_iter = Ytil.TO_calc(nedogrev)[0]  #calc 
stream_posmotret = Ytil.TO_calc(nedogrev)[1]  #calc 
Cond = Condensator(T_cond,stream_reg_cond)   
stream_cond_nasos_iter = Cond.Condensator_calc()
# print(stream_cond_nasos_iter,'После конденсатора')

print(stream_ytil_turb, 'stream_ytil_turb ')
print(stream_cond_nasos, 'stream_cond_nasos') 
print("Начало расчета")
print(stream_turb_reg,' stream_turb_reg')
print(stream_reg_cond, 'stream_reg_cond') 
print(stream_cond_nasos_iter,' stream_cond_nasos_iter')
print(stream_nasos_reg,' stream_nasos_reg')
print(stream_reg_ytil, 'stream_reg_ytil_iter') 
print(stream_ytil_turb_iter,"stream_ytil_turb_iter")


print(stream_posmotret, 'stream_posmotret_outlet_gas')
print(stream_vnesh, 'inlet_gas')
print(stream_cond_nasos['P'])
# print(stream_cond_nasos[1], 'P_end')
```
